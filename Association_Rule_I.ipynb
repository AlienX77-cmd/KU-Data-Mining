# Association Rules: Apriori Algorithm

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```
dataset = [['A','C','D'],
           ['B','C','E'],
           ['A','B','C','E'],
           ['B','E'],
           ['A','B','C','E']
           ]
te = TransactionEncoder()
te_array = te.fit_transform(dataset)
df = pd.DataFrame(te_array,columns=te.columns_)
df
```

```
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```
#my_rules = association_rules(frequent_itemsets,metric="lift",min_threshold=1)
my_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
my_rules
```

```
res = my_rules[my_rules['support'] > 0.6]
res
```
