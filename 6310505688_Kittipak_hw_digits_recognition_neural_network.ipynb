```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

##**Kittipak's Neural Network Code**

###**Given a simple neural network which has only input and output layers, your task is to fine-tune the neural network to obtain your best performance.**

```python
from keras import layers, models

model = models.Sequential([ # using three layers (first layer has 128 nodes, second layer has 64 nodes and third layer has 10 nodes)
    layers.Dense(128, input_shape=(784,), activation='relu'),  # Increased units and changed activation and used relu as the activation function
    layers.Dense(64, activation='relu'),  # Added another hidden layer and use relu as the activation function
    layers.Dense(10, activation='softmax')  # Output layer with softmax for multi-class classification
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=20) # try 20 training cycles
```

```python
# Best Performance that I can obtain from fine-tuning the neural network model

model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
plt.matshow(X_test[0])
```

```python
# Make predictions on the test set
y_predicted_probabilities = model.predict(X_test_flattened)

# Convert predicted probabilities to predicted labels
y_predicted_labels = np.argmax(y_predicted_probabilities, axis=1)
```

```python
#cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
