##**Scatter Plot**

```python
temp = []
for i in ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']:
  for j in ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']:
    if (i == j):
      continue
    temp.append(i)
    if (j in temp): continue
    print(i, j)
    plt.xlabel(i)
    plt.ylabel(j)
    plt.scatter(x=df[i], y=df[j])
    plt.show()
    print("\n")
```

##**3. Processing categorical and/or numerical data**

Categorical to Numerical

```python
data = df.copy()
```

```python
data.tail()
```

```python
data.columns
cat_col = [feature for feature in data.columns if data[feature].dtypes=='O' and feature != "income"]
cat_col
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

#data['relationship']=encoder.fit_transform(data['relationship'])
#data

for feature in cat_col:
  data[feature] = encoder.fit_transform(data[feature])
```

```python
data.tail()
```

####**Note ทำการแยก income ออกไปเพราะมันเป็น target variable แล้วเด่วค่อย concat หรือเพิ่ม column income เข้าไปที่ dataframe ทีหลัง หลังจาก scaling ไปแล้ว**

```python
tmp = data["income"]
tmp
```
